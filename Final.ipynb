```python
# Data collection: OOI Regional Cabled Array Axial Seamount (Cruise Data)
# https://dataexplorer.oceanobservatories.org/#platform/fffe2293-19ad-55f4-83dd-1e97b5854063/v2?pid=23&tab=visualization
```

```python
# Import statements
import pandas as pd
import matplotlib.pyplot as plt

# Importing my data and getting variables
df = pd.read_csv('ooi.csv', delimiter=',')
temp = df['ctd_temperature_1'].values
con= df['ctd_conductivity_1'].values
oxygen = df['ctd_oxygen'].values
depth = df['depth'].values

# Creating my figures
plt.figure()
fig,(ax0,ax1,ax2) = plt.subplots(ncols=3, figsize=(20,10))

# Plot 1: Temperature vs Depth
ax0.scatter(temp,depth,c='mediumblue',s=70)
ax0.set_title('Temperature (C) vs Depth (m)',fontsize=18)
ax0.set_xlabel('Temperature (C)',fontsize=15)
ax0.set_ylabel('Depth (m)',fontsize=15)
ax0.invert_yaxis()
ax0.grid()

# Plot 2: Conductivity vs Depth
ax1.scatter(con,depth,c='orangered',s=70)
ax1.set_title('Conductivity (mS/cm) vs Depth (m)',fontsize=18)
ax1.set_xlabel('Conductivity (mS/cm)',fontsize=15)
ax1.set_ylabel('Depth (m)',fontsize=15)
ax1.invert_yaxis()
ax1.grid()

# Plot 3: Oxygen Saturation vs Depth
ax2.scatter(oxygen,depth,c='teal',s=70)
ax2.set_title('Oxygen (mol/L) vs Depth (m)',fontsize=18)
ax2.set_xlabel('Oxygen (mol/L)',fontsize=15)
ax2.set_ylabel('Depth (m)',fontsize=15)
ax2.invert_yaxis()
ax2.grid()
```
